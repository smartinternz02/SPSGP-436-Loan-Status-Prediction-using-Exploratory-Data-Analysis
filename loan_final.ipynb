```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('credit_train.csv')
```

```python
data
```

```python
for i in range(100000 ,100514):
    data = data.drop([i])
```

```python
data['Home Ownership'].unique()
```

```python
data.isnull().any()
```

```python
data['Bankruptcies'].fillna(0,inplace=True)
data['Tax Liens'].fillna(data['Tax Liens'].mode()[0],inplace=True)
data['Credit Score'].fillna(data['Credit Score'].mode()[0],inplace=True)
data['Annual Income'].fillna(data['Annual Income'].mean(),inplace=True)
data['Years in current job'].fillna(data['Years in current job'].mode()[0],inplace=True)
data['Maximum Open Credit'].fillna(data['Maximum Open Credit'].mode()[0],inplace=True)
data['Tax Liens'].fillna(data['Tax Liens'].mode()[0],inplace=True)
```

```python
data.drop(['Loan ID','Customer ID','Purpose','Monthly Debt','Months since last delinquent','Number of Open Accounts'],axis=1,inplace=True)
```

```python
data.isnull().any()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='Term',data=data)
```

```python
sns.countplot(x='Loan Status',data=data)
```

```python
data['Years in current job'] = data['Years in current job'].map({'8 years':8, '10+ years':15,'3 years':3, '5 years':5, '< 1 year':0.5, '2 years':2, '4 years':4, '9 years':9, '7 years':7, '1 year':1, '6 years':6})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Term'] = le.fit_transform(data['Term'])
data['Home Ownership'] = le.fit_transform(data['Home Ownership'])
```

```python
data.head()
```

```python
x = data.iloc[:,1:13].values
y = data.iloc[:,0:1].values
```

```python
x.shape
```

```python
x[0]
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
z = one.fit_transform(x[:,5:6]).toarray()
x = np.delete(x,1,axis=1)
x = np.concatenate((z,x),axis=1)
```

```python
x[0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=0)
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
y_pred
```

```python
y_test
```

```python
y_test.size
```

```python
y_pred.size
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
```
